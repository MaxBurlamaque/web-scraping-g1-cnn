Sugestões de uso:

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
```

# Web Scraping - g1 - politica

```python
driver_g1 = webdriver.Chrome(ChromeDriverManager().install())
url_g1 = 'https://g1.globo.com/politica/'
driver_g1.get(url_g1)
soup_g1 = BeautifulSoup(driver_g1.page_source, 'html.parser')
```

```python
materias = soup_g1.find_all('div', {'data-type': 'materia'})
```

```python
for i in materias:
    print("\n\n---")
    print("link:")
    print(i.h2.a.get('href'))
    
    print('\ntitulo:')
    print(i.h2.a.text)
    
    print('\ndescriçao da materia:')
    try:
        print(i.find_all('div', {"class": 'feed-post-body-resumo'})[0].text)
    except:
        print("nulo")
        pass
    print('\nDATA:')
    print(i.find_all('div', {'class': 'feed-post-metadata'})[0].find_all('span', {'feed-post-datetime'})[0].text)
    print('\nSECTION:')
    print(i.find_all('div', {'class': 'feed-post-metadata'})[0].find_all('span', {'feed-post-metadata-section'})[0].text)
    
    
    url_n = i.h2.a.get('href')
    driver_g1.get(url_n)
    soup_g1 = BeautifulSoup(driver_g1.page_source, 'html.parser')
    
    try:
        results2 = soup_g1.find('div', {'class': 'mc-column entities'})
        print("\ntags:")
        print(results2.li.find('a').get("data-track-click"))
    except:
        pass
    
    
    results3 = soup_g1.find('div', {'class': 'content-publication-data'})
    print('\nAutor:')
    print(results3.find('span', {'itemprop': 'author'}).find('meta').get('content'))
    print('\ndatas:')
    print(results3.find('p', {"class": "content-publication-data__updated"}).time.get('datetime'))
    print(results3.find('p', {"class": "content-publication-data__updated"}).span.time.get('datetime'))
    print(results3.find('p', {"class": "content-publication-data__updated"}).time.text)
    
    
    try:
        results4 = soup_g1.find('p', {'class': 'content-publication-data__from'})
        print('\nidentificaçoes: (nem sempre tem)')
        print(results4.text)
    except:
        pass
    
    try:
        print('\n', soup_g1.find('h2', {"class": "content-head__subtitle"}).text)
    except:
        pass
    
    res1 = soup_g1.find_all('div', {'class': 'mc-column content-text active-extra-styles'})
    for i in res1:
        print('\n', i.text)
    
    
    
    driver_g1.get(url_g1)
    soup_g1 = BeautifulSoup(driver_g1.page_source, 'html.parser')    
    break # Comentar esse break para nao ficar spamado, mas ele passa por todas noticias da pagina principal de politica
```

# Web Scraping - cnn - politica

```python
driver_cnn = webdriver.Chrome(ChromeDriverManager().install())
url_cnn = 'https://www.cnnbrasil.com.br/politica/'
driver_cnn.get(url_cnn)
soup_cnn = BeautifulSoup(driver_cnn.page_source, 'html.parser')
```

```python
materias_cnn = soup_cnn.find('ul', {'class': 'home__new'}).find_all('li')
```

```python
for i in materias_cnn:
    
    #NA PAGINA DE RECENTES
    
    print("\n\n---")
    print("\nlink:")
    print(i.a.get('href'))
    
    print('\ntitulo:')
    print(i.a.text.strip())
    
    print('\ndescriçao da materia:')
    print("diferente do g1, a cnn nao tem descriçao explicita na pagina de novas noticias")
    print('\nDATA:')
    print(i.find('div', {'class': 'latest__news__infos'}).span.text.strip())
    
    print('\nSECTION:')
    print(i.find('div', {'class': 'latest__news__infos'}).a.text)
    
    
    # DENTRO DA PAGINA
    
    
    url_nx = i.a.get('href')
    driver_cnn.get(url_nx)
    soup_cnn = BeautifulSoup(driver_cnn.page_source, 'html.parser')
    
    
    
    
    print("\ntitulo:")
    print(soup_cnn.find('header', {'class': 'post__header'}).find('h1', {'class', 'post__title'}).text.strip())
    
    print('\nsubtitulo:')
    print(soup_cnn.find('header', {'class': 'post__header'}).find('p', {'class', 'post__excerpt'}).text)
    
    print("\nautor:")
    print(soup_cnn.find('header', {'class': 'post__header'}).find('p', {'class', 'author__name'}).text)
    
    print('\ndata:')
    print(soup_cnn.find('header', {'class': 'post__header'}).find('span', {'class', 'post__data'}).text )
    
    print("\ntags:")
    print(soup_cnn.find('ul', {'class': 'tags__list'}).text)
    
    #body
    for i in soup_cnn.find('div', {'class': 'post__content'}).find_all('p'):
        print(i.text)
   
    
    driver_cnn.get(url_cnn)
    soup_cnn = BeautifulSoup(driver_cnn.page_source, 'html.parser')    
    
    break
```
